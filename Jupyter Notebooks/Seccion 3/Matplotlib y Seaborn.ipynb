# Introducción

A veces puede parecer más fácil pasar por un conjunto de puntos de datos y obtener información a partir de ellos, pero por lo general este proceso puede no producir buenos resultados. Podría haber muchas cosas sin descubrir como resultado de este proceso. Además, la mayoría de los conjuntos de datos utilizados en la vida real son demasiado grandes para realizar un análisis manualmente. Aquí es esencialmente donde interviene la visualización de datos.

La visualización de datos es una forma más fácil de presentar los datos, por compleja que sea, para analizar tendencias y relaciones entre variables con la ayuda de la representación pictórica.

Las siguientes son las ventajas de la visualización de datos:

* Representación más sencilla de datos.
* Destaca las áreas de buen y mal desempeño.
* Explora la relación entre puntos de datos.
* Identifica patrones de datos incluso para puntos de datos más grandes.

Al crear la visualización, siempre es una buena práctica tener en cuenta algunos de los puntos mencionados a continuación.

* Garantizar el uso adecuado de formas, colores y tamaños al crear la visualización.
* Los diagramas / gráficos que utilizan un sistema de coordenadas son más pronunciados.
* El conocimiento de la trama adecuada con respecto a los tipos de datos aporta más claridad a la información.
* El uso de etiquetas, títulos, leyendas y consejos transmite información fluida a una audiencia más amplia.

Existen mucha librería para visualización de datos en Python, sin embargo existen dos que destacan por su fácil uso y además son las herramientas más usadas en machine learning, estan son **Matplotlib y Seaborn**.

# Naturaleza de la visualización

Dependiendo de la cantidad de variables utilizadas para trazar la visualización y el tipo de variables, podría haber diferentes tipos de gráficos que podríamos usar para comprender la relación. Según el recuento de variables, podríamos tener:

* Gráfico univariado(involucra solo una variable)
* Gráfico bivariado(se requiere más de una variable)

Una **gráfica univariante** podría ser para una **variable continua** para comprender la extensión y distribución de la variable, mientras que para una **variable discreta** podría decirnos el recuento.

De manera similar, una **gráfica bivariada** para una **variable continua** podría mostrar una estadística esencial como una correlación, ya que una variable continua versus una discreta podría llevarnos a conclusiones muy importantes, como comprender la distribución de datos en diferentes niveles de una variable categórica. También se podría desarrollar una gráfica bivariada entre dos variables discretas.

## Diagrama de caja

Un **diagrama de caja**, también conocido como diagrama de caja y bigotes, la caja y el bigote se muestran claramente en la imagen de abajo. Es una muy buena representación visual cuando se trata de medir la distribución de datos. Traza claramente los valores medianos, los valores atípicos y los cuartiles. Comprender la distribución de datos es otro factor importante que conduce a una mejor construcción de modelos. Si los datos tienen valores atípicos, el diagrama de caja es una forma recomendada de identificarlos y tomar las acciones necesarias.

El gráfico de caja y bigotes muestra cómo se distribuyen los datos. En general, se incluyen cinco piezas de información en el gráfico.

* El mínimo se muestra en el extremo izquierdo del gráfico, al final del 'bigote' izquierdo
* El primer cuartil, Q1, es el extremo izquierdo del cuadro(bigote izquierdo)
* La mediana se muestra como una línea en el centro del cuadro
* Tercer cuartil, Q3, que se muestra en el extremo derecho del cuadro(bigote derecho)
* El máximo está en el extremo derecho del cuadro.

### Sintáxis

### Ejemplo

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("data/diabetes.csv")
df.head()
```

```python
fig = plt.figure(1, figsize=(9, 6))
plt.boxplot(df['BloodPressure'])
plt.title('Diagrama de caja')
plt.show()
```

* Los valores mínimo y máximo se consideran excluyendo los valores atípicos en el conjunto de datos.

* Vea ese recuadro en el medio, ya que la parte inferior denota el percentil 25, es decir, el primer cuartil, y la parte superior representa el percentil 75, es decir, el tercer cuartil.

* La línea naranja del medio representa el valor mediano.

* La distancia entre la línea de cuantil 1 y 3 se denomina rango entre cuartiles.

* **¿Qué es el percentil?** Si un punto tiene un percentil de 40, significa que el 40% de los puntos del conjunto de datos están por debajo de este.

### Curvas de Nivel

La función matplotlib.contour (o su método equivalente) nos permite mostrar en una gráfica curvas de nivel de una forma muy sencilla. Los valores a pasar como argumentos son arrays bidimensionales con los valores x e y de los puntos (obtenidos con la función numpy.meshgrid) y un array, de las mismas dimensiones que x e y, con los valores en el eje z a representar.

```python
import numpy as np

def f(x, y):
    return np.sin(x) ** 2 + np.cos(5 + x * y) + 2 * np.cos(x)
    
x = np.linspace(0, 5, 100)
y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X,Y,Z)
plt.show()
```

#### Agregando niveles

```python
plt.contour(X,Y,Z,levels=15)
plt.show()
```

## Graficos 3D

Para habilitar la creación de este tipo de gráficos basta importar el objeto Axes3D de la sublibrería mpl_toolkits.mplot3d, y generar un conjunto de ejes con el método gca asociado a la figura indicando que la proyección es de tipo "3d":

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure()
fig.gca(projection='3d')
plt.show()
```

### Superficies
Una vez creado el conjunto de ejes 3D, mostrar una superficie tridimensional en estos ejes es relativamente sencillo usando el método del conjunto de ejes plot_surface. Este método recibe como argumentos de entrada arrays 2D con los valores x, y y z de la misma manera que vimos para la función contour. Por ejemplo, partamos del siguiente conjunto de puntos:

```python
X = np.arange(-10, 10, 0.25)
Y = np.arange(-10, 10, 0.25)
X, Y = np.meshgrid(X, Y)
Z = np.sin(np.sqrt(X**2 + Y**2))
```

```python
fig=plt.figure()
ax=fig.gca(projection='3d')
surface=ax.plot_surface(X,Y,Z)
plt.show()
```

### Graficos de malla (Wireframes)

El método es plot_wireframe y recibe como primeros tres argumentos las matrices X, Y y Z que hemos visto para la función plot_surface. En este caso, el método no es compatible con los mapas de color, por lo que no es posible asignarle uno ni crear una barra de color que acompañe al gráfico:

```python
fig=plt.figure()
ax=fig.gca(projection='3d')
surface=ax.plot_wireframe(X,Y,Z)
plt.show()
```

### Curvas de contornos
El método contour3D nos permite crear gráficos de curvas de nivel en 3D. En su configuración más básica simplemente recibe como argumentos los arrays X, Y y Z:

```python
fig=plt.figure()
ax=fig.gca(projection='3d')
surface=ax.contour3D(X,Y,Z)
plt.show()
```

### Gráficos de dispersión 3d

También podemos crear gráficos de dispersión fácilmente con el método scatter3D. Una vez más, los tres argumentos principales son los arrays X, Y y Z conteniendo la información sobre los puntos a mostrar. En el siguiente ejemplo generamos dichos datos de forma aleatoria:

```python
fig=plt.figure(figsize=(14,6))
ax=fig.gca(projection='3d')
for n in range(3):
    x=np.random.normal(0,(n+1)*3, 100)
    y=np.random.normal(0,(n+1)*3, 100)
    z=np.random.normal(0,(n+1)*3, 100)
    scatter=ax.scatter3D(x,y,z,label=n)
plt.legend()
plt.show()
```
